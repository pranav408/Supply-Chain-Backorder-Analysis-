ttest

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the dataset
df = pd.read_csv('/content/backorders-data-set.csv')

# Assuming 'national_inv' is the variable of interest
# Extract data for the two groups (backorder vs. non-backorder)
group_backorder = df[df['went_on_backorder'] == 1]['national_inv']
group_non_backorder = df[df['went_on_backorder'] == 0]['national_inv']

# Perform two-sample t-test
statistic, p_value = ttest_ind(group_backorder, group_non_backorder)

# Print the test statistic and p-value
print("Two-sample t-test Statistic:", statistic)
print("P-value:", p_value)

# Interpretation of results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")
```

power analysis of ttest

```python
import numpy as np
from statsmodels.stats.power import TTestIndPower
import matplotlib.pyplot as plt

# Parameters for power analysis
effect_sizes = np.arange(0.1, 1.0, 0.1)  # Effect sizes to test
alpha = 0.05  # Significance level
nobs1 = 100  # Sample size for group 1 (backorder)
nobs2 = 100  # Sample size for group 2 (non-backorder)

# Initialize power analysis object
power_analysis = TTestIndPower()

# Calculate power for different effect sizes
powers = power_analysis.solve_power(effect_size=effect_sizes, nobs1=nobs1, alpha=alpha)

# Plot power curve
plt.plot(effect_sizes, powers, marker='o')
plt.title('Power Analysis for Two-Sample t-Test')
plt.xlabel('Effect Size')
plt.ylabel('Power')
plt.show()
```

Wilcoxon signed-rank test

```python
from scipy.stats import wilcoxon
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/backorders-data-set.csv')

# Assuming 'national_inv' and 'went_on_backorder' are the variables of interest
# Extract the variables
national_inv = df['national_inv']
went_on_backorder = df['went_on_backorder']

# Perform the Wilcoxon signed-rank test
statistic, p_value = wilcoxon(national_inv[went_on_backorder == 0], national_inv[went_on_backorder == 1])

# Set the significance level (alpha)
alpha = 0.05

# Print the test statistic and p-value
print("Wilcoxon Signed-Rank Test Statistic:", statistic)
print("P-value:", p_value)

# Interpret the results
if p_value > alpha:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude a significant difference.")
else:
    print("Reject the null hypothesis: There is sufficient evidence to conclude a significant difference.")
```

power analysis on wilcoxon signed-rank test

```python
from scipy.stats import wilcoxon
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
effect_sizes = np.arange(0.1, 1.0, 0.1)  # Effect sizes to test
sample_size = 100  # Sample size
alpha = 0.05  # Significance level
num_simulations = 1000  # Number of simulations for each scenario

# Function to perform power analysis for a specific effect size
def power_analysis(effect_size):
    # Initialize counter for rejections
    num_rejections = 0

    # Simulate data and perform Wilcoxon signed-rank test num_simulations times
    for _ in range(num_simulations):
        # Generate data for two groups with specified effect size
        group1 = np.random.normal(loc=0, scale=1, size=sample_size)
        group2 = np.random.normal(loc=effect_size, scale=1, size=sample_size)

        # Perform Wilcoxon signed-rank test
        statistic, p_value = wilcoxon(group1, group2)

        # Check if null hypothesis is rejected
        if p_value < alpha:
            num_rejections += 1

    # Calculate power as proportion of rejections
    power = num_rejections / num_simulations
    return power

# Perform power analysis for each effect size
powers = [power_analysis(effect_size) for effect_size in effect_sizes]

# Plot power analysis graph
plt.figure(figsize=(10, 6))
plt.plot(effect_sizes, powers, marker='o')
plt.title('Power Analysis for Wilcoxon Signed-Rank Test')
plt.xlabel('Effect Size')
plt.ylabel('Power')
plt.grid(True)
plt.show()
```

comparing power curves of 2 sample test and Wilcoxon signed rank-test

```python
from scipy.stats import ttest_ind, wilcoxon
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
effect_sizes = np.arange(0.1, 1.0, 0.1)  # Effect sizes to test
sample_size = 50  # Sample size
alpha = 0.05  # Significance level
num_simulations = 1000  # Number of simulations for each scenario

# Function to perform power analysis for two-sample t-test
def ttest_power(effect_size):
    # Initialize counter for rejections
    num_rejections = 0

    # Simulate data and perform two-sample t-test num_simulations times
    for _ in range(num_simulations):
        # Generate data for two groups with specified effect size
        group1 = np.random.normal(loc=0, scale=1, size=sample_size)
        group2 = np.random.normal(loc=effect_size, scale=1, size=sample_size)

        # Perform two-sample t-test
        _, p_value = ttest_ind(group1, group2)

        # Check if null hypothesis is rejected
        if p_value < alpha:
            num_rejections += 1

    # Calculate power as proportion of rejections
    power = num_rejections / num_simulations
    return power

# Function to perform power analysis for Wilcoxon signed-rank test
def wilcoxon_power(effect_size):
    # Initialize counter for rejections
    num_rejections = 0

    # Simulate data and perform Wilcoxon signed-rank test num_simulations times
    for _ in range(num_simulations):
        # Generate data for two groups with specified effect size
        group1 = np.random.normal(loc=0, scale=1, size=sample_size)
        group2 = np.random.normal(loc=effect_size, scale=1, size=sample_size)

        # Perform Wilcoxon signed-rank test
        _, p_value = wilcoxon(group1, group2)

        # Check if null hypothesis is rejected
        if p_value < alpha:
            num_rejections += 1

    # Calculate power as proportion of rejections
    power = num_rejections / num_simulations
    return power

# Perform power analysis for each effect size
ttest_powers = [ttest_power(effect_size) for effect_size in effect_sizes]
wilcoxon_powers = [wilcoxon_power(effect_size) for effect_size in effect_sizes]

# Plot power analysis graph
plt.figure(figsize=(10, 6))
plt.plot(effect_sizes, ttest_powers, marker='o', label='Two-Sample t-Test')
plt.plot(effect_sizes, wilcoxon_powers, marker='o', label='Wilcoxon Signed-Rank Test')
plt.title('Power Analysis Comparison')
plt.xlabel('Effect Size')
plt.ylabel('Power')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, wilcoxon

# Define parameters
effect_sizes = np.arange(0.1, 1.0, 0.1)  # Effect sizes to test
sample_size = 50  # Sample size
alpha = 0.05  # Significance level
num_simulations = 1000  # Number of simulations for each scenario

# Initialize lists to store the p-values for each effect size
ttest_p_values = []
wilcoxon_p_values = []

# Perform tests for each effect size
for effect_size in effect_sizes:
    # Initialize arrays to store p-values for each simulation
    ttest_p_values_sim = np.zeros(num_simulations)
    wilcoxon_p_values_sim = np.zeros(num_simulations)

    # Simulate data and perform tests num_simulations times
    for i in range(num_simulations):
        # Generate data for two groups with specified effect size
        group1 = np.random.normal(loc=0, scale=1, size=sample_size)
        group2 = np.random.normal(loc=effect_size, scale=1, size=sample_size)

        # Perform two-sample t-test
        _, ttest_p_values_sim[i] = ttest_ind(group1, group2)
        # Perform Wilcoxon signed-rank test
        _, wilcoxon_p_values_sim[i] = wilcoxon(group1, group2)

    # Append median p-values to the lists
    ttest_p_values.append(np.median(ttest_p_values_sim))
    wilcoxon_p_values.append(np.median(wilcoxon_p_values_sim))

# Plot p-values for t-test
plt.figure(figsize=(10, 6))
plt.plot(effect_sizes, ttest_p_values, marker='o', label='Two-Sample t-Test')
plt.axhline(y=alpha, color='r', linestyle='--', label='Significance Level')
plt.title('Median p-values for Two-Sample t-Test across Effect Sizes')
plt.xlabel('Effect Size')
plt.ylabel('Median p-value')
plt.legend()
plt.grid(True)
plt.show()

# Plot p-values for Wilcoxon signed-rank test
plt.figure(figsize=(10, 6))
plt.plot(effect_sizes, wilcoxon_p_values, marker='o', label='Wilcoxon Signed-Rank Test')
plt.axhline(y=alpha, color='r', linestyle='--', label='Significance Level')
plt.title('Median p-values for Wilcoxon Signed-Rank Test across Effect Sizes')
plt.xlabel('Effect Size')
plt.ylabel('Median p-value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def normal_plot(data, title='Normal Plot'):
    plt.figure(figsize=(8, 6))

    # Plot histogram for data
    plt.hist(data, bins=20, density=True, alpha=0.6, color='b', label='Data')

    # Fit a normal distribution to the data
    mu, std = norm.fit(data)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')

    # Add labels and title
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage for unpaired sample situation
data_unpaired = np.random.normal(loc=0, scale=1, size=100)  # Generating unpaired data
normal_plot(data_unpaired, title='Normal Plot of Data')
```
